a. Addition Rule:

```python
# Addition Rule Example

# P(A) = 0.3, P(B) = 0.4, P(A  ‚à© B) = 0.1

p_a = 0.3  # Probability of event A
p_b = 0.4  # Probability of event B
p_a_and_b = 0.1  # Probability of both A and B

# P(A ‚à™ B) = P(A) + P(B) - P(A ‚à© B)
p_a_or_b = p_a + p_b - p_a_and_b
print(f"Probability of A or B (Addition Rule): {p_a_or_b}")
```

b. Multiplication Rule:

```python
# Multiplication Rule Example

# P(A) = 0.5, P(B) = 0.4

p_a = 0.5
p_b = 0.4

# P(A ‚à© B) = P(A) * P(B)
p_a_and_b = p_a * p_b
print(f"Probability of A and B (Multiplication Rule): {p_a_and_b}")
```

c. Conditional Probability:

```python
# Conditional Probablity Example

# P(A ‚à© B) = 0.2, P(B) = 0.5

p_a_and_b = 0.2
p_b = 0.5

# P(A | B) = P(A ‚à© B) / P(B)
p_a_given_b = p_a_and_b / p_b
print(f"Conditional Probability of A given B: {p_a_given_b}")
```

# 2. Probability Distributions

a. Normal Distribution:

What is a Normal Distribution?

- The normal distribution is a probability distribution that is symmetric and bell-shaped. Most of the data points are concentrated around the mean, and the probabilities decrease as you move further away from the mean. It's commonly used in statistics because many real-world phenomena (e.g., heights, test scores) follow a normal distribution.

Key properties:

- Mean (¬µ): The center of the distribution.

Standard deviation (œÉ): Determines the spread of the data. A smaller œÉ means data is more concentrated around the mean.

It's symmetric about the mean.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step1: Define the mean(center) and standard deviation (spread)
mean = 0        # Center of the curve
std_dev = 1    # Spread of the rcurve

# Step2: Generate x-values (range of values around the mean)
x = np.linspace(-4, 4, 100)   # Generate 100 points between -4 and 4

# Step3: Calculate the y-labels (height of the curve) using the formula of normal ditribution
# y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
y = norm.pdf(x, mean, std_dev)

# Step4: plot the curve
plt.plot(x, y)
plt.title("Normal Distribution (Mean = 0, Standard Deviation = 1)")
plt.xlabel("X - axis (Values)")
plt.ylabel("Y - axis (Probability Density)")
plt.grid()
plt.show()
```

b. Bionomial distribution

Definition of Binomial Distribution:
- A binomial distribution is used to find the probability of a specific number of successes in a fixed number of trials, where each trial has only two outcomes: success or failure (like flipping a coin).

- It answers questions like:
"What is the probability of getting exactly 3 heads in 10 coin flips?"

Key Points:
1. Fixed number of trials (n): You repeat an experiment a certain number of times.
2. Two outcomes: Success (e.g., heads) or failure (e.g., tails).
3. Constant probability (p): The probability of success stays the same in every trial.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Step1: Define the paramethers
n = 10    # Total number of trails (e.g., 10 coin flips)
p = 0.5   # Probability of success (e.g., probability of heads)

# Step2: Define the range of possible outcomes (0 to n)
x = range(n + 1)   # Possible number of success (0, 1, ...., 10)

# Step3: Calculate probabilities for each possible outcome
probabilities = binom.pmf(x, n, p)   # PMF gives the probability for each x

# Step4: Plot the probabilities
plt.bar(x, probabilities, color = 'lightblue', edgecolor = 'black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Success (Heads)")
plt.ylabel("Probability")
plt.grid(axis ='y', linestyle = '-', alpha = 0.7)
plt.show()
```

c. Poisson Distribution

A Poisson distribution is a discrete probability distribution. It gives the probability of an event happening a certain number of times (k) within a given interval of time or space. The Poisson distribution has only one parameter, Œª (lambda), which is the mean number of events.

Imagine a small coffee shop. On average, they receive 5 customers per hour. We can use the Poisson distribution to calculate the probability of receiving a certain number of customers in a given hour.


Formula:

The probability of observing k events in an interval is given by:

P(X = k) = (Œª^k * e^(-Œª)) / k!

```python
import math
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define parameters
lambda_rate = 3  # Average rate (emails per hours)
k = 5

# Step 2: Poisson formula to calculate probability
probability = poisson.pmf(k, lambda_rate)

# Poisson formuls
P = (lambda_rate**k * math.exp(-lambda_rate)) / math.factorial(k)

print(f"The probability of receiving exactily 5 emails is: {P: .4f}")
print(f"The probability of receiving exactly 5 emails in an hour is {probability:.4f}")

# Step 3: Generate probabilities for k = 0 to 10
x = range(0, 11)   # Possible number of emails (0 to 10)
probabilities = [(lambda_rate**i * math.exp(-lambda_rate)) / math.factorial(i) for i in x]

# Step 4: Plot the distribution
plt.bar(x, probabilities, color = 'lightgreen', edgecolor = 'black')
plt.title(f"Poisson Distribution (Œª = {lambda_rate})")
plt.xlabel("Number of Emials (k)")
plt.ylabel("Probability")
plt.axhline(y=P, color='red', linestyle='--', label=f"P(X = {k}) = {P:.4f}")
plt.legend()
plt.show()
```

d. Uniform Distribution

A uniform distribution is a type of probability distribution in which every event has the same probability of occurring within a given range.

In simpler terms, if we have a uniform distribution, the outcomes are equally likely to happen. For example, if we roll a fair die, the probability of rolling any of the six numbers (1 through 6) is the same, so it's a uniform distribution.

Key Characteristics:
Continuous Uniform Distribution: The values are within a range, and any value in that range is equally likely.
For example, picking a random number between 0 and 1 (i.e., any decimal between 0 and 1 has the same chance of being chosen).
Discrete Uniform Distribution: The values are distinct, and each has the same chance of being chosen.
For example, a fair die has 6 faces, each equally likely to show up.

Python Code for Uniform Distribution (Discrete):

Here‚Äôs a simple Python example to visualize a discrete uniform distribution where we randomly pick an integer between 1 and 6 (like a fair die):

Uniform Distribution (Discrete):

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define parameters
a, b = 1, 6  # Range of possible outcomes (1 to 6)

# Step 2: Generate data
outcomes = np.arange(a , b+1)  # Possible outcomes: 1, 2, 3, 4, 5, 6
probabilities = np.ones_like(outcomes) / (b - a + 1)   #  Equal probability for each outcome

# Step 3: Plot the distribution
plt.bar(outcomes, probabilities, color='lightgreen', edgecolor='black')
plt.title("Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks(outcomes)
plt.ylim(0, 1)
plt.show()
```

Uniform Distribution (Continuous):

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define parameters
a, b = 0, 1  # Range of possible outcomes (0 to 1)

# Step 2: Generate data
x = np.linspace(a, b, 1000)  # Generate 1000 points between 0 and 1
pdf = np.ones_like(x) / (b - a)  # Equal probability density for each point

# Step 3: Plot the distribution
plt.plot(x, pdf, color='blue')
plt.title("Continuous Uniform Distribution (0 to 1)")
plt.xlabel("Outcome")
plt.ylabel("Probability Density")
plt.ylim(0, 1.2)
plt.show()
```
