1. Measures of Central Tendency

- Mean: The average of the data.
- Median: The middle value when data is sorted.
- Mode: The most frequently occurring value.

```python
# in series
import pandas as pd

# Sample dataset: Heights of 10 people (in cm)
data = [150, 160, 165, 170, 155, 180, 175, 165, 160, 170]

# Convert data into a pandas Series
data_series = pd.Series(data)

# Calculating measures of central tendency
mean = data_series.mean()  # Mean
median = data_series.median()  # Median
mode = data_series.mode()[0]  # Mode

print("Measures of Central Tendency:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

```python
# in dataframe
import pandas as pd

# Sample dataset: Heights of 10 people (in cm)
data = [150, 160, 165, 170, 155, 180, 175, 165, 160, 170]

# Convert data into a pandas DataFrame
df = pd.DataFrame(data, columns=['Height'])

# Calculating measures of central tendency
mean = df['Height'].mean()  # Mean
median = df['Height'].median()  # Median
mode = df['Height'].mode()[0]  # Mode

print("Measures of Central Tendency:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

2. Measures of Spread
- Variance: How far the data points are from the mean.
- Standard Deviation: Square root of variance.
- Range: Difference between max and min values.
- IQR: Range between 25th percentile (Q1) and 75th percentile (Q3).

```python
# in series
# Calculating measures of spread
variance = data_series.var()  # Variance
std_dev = data_series.std()  # Standard Deviation
range_value = data_series.max() - data_series.min()  # Range
q1 = data_series.quantile(0.25)  # 25th percentile
q3 = data_series.quantile(0.75)  # 75th percentile
iqr = q3 - q1  # Interquartile Range

print("\nMeasures of Spread:")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Range: {range_value}")
print(f"IQR: {iqr}")
```

```python
# in dataframe
import pandas as pd

# Sample dataset: Heights of 10 people (in cm)
data = [150, 160, 165, 170, 155, 180, 175, 165, 160, 170]

# Convert data into a pandas DataFrame
df = pd.DataFrame(data, columns=['Height'])

# Calculating measures of spread
variance = df['Height'].var()  # Variance
std_dev = df['Height'].std()  # Standard Deviation
range_value = df['Height'].max() - df['Height'].min()  # Range
q1 = df['Height'].quantile(0.25)  # 25th percentile
q3 = df['Height'].quantile(0.75)  # 75th percentile
iqr = q3 - q1  # Interquartile Range

print("\nMeasures of Spread:")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Range: {range_value}")
print(f"IQR: {iqr}")
```

3. Data Visualization
- a. Histogram:
Shows the distribution of the data.

```python
# in series
import matplotlib.pyplot as plt

# Histogram
plt.figure(figsize=(6, 4))
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset: Heights of 10 people (in cm)
data = [150, 160, 165, 170, 155, 180, 175, 165, 160, 170]

# Convert data into a pandas DataFrame
df = pd.DataFrame(data, columns=['Height'])

# Histogram
plt.figure(figsize=(6, 4))
plt.hist(df['Height'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```

b. Box Plot:
Displays spread, median, and outliers.

```python
# in series
import seaborn as sns

# Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, color='lightgreen')
plt.title('Box Plot')
plt.ylabel('Height')
plt.show()
```

```python
# in dataframe
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset: Heights of 10 people (in cm)
data = [150, 160, 165, 170, 155, 180, 175, 165, 160, 170]

# Create a DataFrame for Seaborn
import pandas as pd
df = pd.DataFrame(data, columns=['Height'])

# Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(data=df['Height'], color='lightgreen')
plt.title('Box Plot')
plt.ylabel('Height')
plt.show()
```

c. Scatter Plot:
Shows relationships between two variables.

```python
# Scatter Plot: Index vs Height
plt.figure(figsize=(6, 4))
plt.scatter(range(len(data)), data, color='orange')
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Height')
plt.show()
```
